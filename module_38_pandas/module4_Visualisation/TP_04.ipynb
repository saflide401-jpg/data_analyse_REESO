# TP-04 : Diagrammes en barres - Budget de l'Etat
##### Partie A : Donnees budgetaires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Création du dictionnaire avec les données
donnees = {
    'Secteur': [
        'Education', 
        'Sante', 
        'Defense et Securite', 
        'Infrastructures', 
        'Agriculture', 
        'Administration', 
        'Autres'
    ],
    'Allocation_Budgetaire (Milliards FCFA)': [
        450, 
        280, 
        350, 
        320, 
        180, 
        200, 
        220
    ]
}

# Création du DataFrame
df_budget = pd.DataFrame(donnees)

# Affichage du DataFrame
print(df_budget)
```

### 2. Calculez le budget total et les pourcentages.

```python
# Calcul du budget total
budget_total = df_budget['Allocation_Budgetaire (Milliards FCFA)'].sum()

#          pourcentages 
df_budget['Pourcentage (%)'] = (df_budget['Allocation_Budgetaire (Milliards FCFA)'] / budget_total) * 100




# Affichage des résultats
print(f"Le budget total est de : {budget_total} Milliards FCFA\n")
print(df_budget)
```

# Partie B : Barres verticales

#### Creez un diagramme en barres verticales montrant le budget par secteur.

```python

#   la taille de la figure

plt.figure(figsize=(12, 6))
plt.bar(df_budget['Secteur'], df_budget['Allocation_Budgetaire (Milliards FCFA)'], color='blue')  # Créer le diagramme en barres

#   titres et labels
plt.title('Répartition du Budget par Secteur (en Milliards FCFA)', fontsize=14, color= 'blue')
plt.xlabel('Secteurs', fontsize=12)
plt.ylabel('Budget (Milliards FCFA)', fontsize=12)

#  Faire pivoter les noms des secteurs pour une meilleure lisibilité
plt.xticks(rotation=45)

#   grille horizontale 
plt.grid(axis='y', linestyle='--',alpha=0.7 )

#  Afficher le graphique
plt.tight_layout()
plt.show()
```

#### 4. Triez les barres par ordre decroissant de budget.

```python


# Trier le DataFrame 
df_sorted = df_budget.sort_values(by='Allocation_Budgetaire (Milliards FCFA)', ascending=False)

# figure configuration
plt.figure(figsize=(12, 6))

# Création du graphique
plt.bar(df_sorted['Secteur'], df_sorted['Allocation_Budgetaire (Milliards FCFA)'], 
        color='steelblue')

# Titres et Labels
plt.title('Budgets par Secteur (Triés par ordre decroissant)', fontsize=15)
plt.ylabel('Milliards FCFA', fontsize=12)
plt.xticks(rotation=45)



plt.tight_layout()
plt.show()
```

### 5. Ajoutez la valeur exacte au-dessus de chaque barre.

```python
 

#  Configuration de la figure
plt.figure(figsize=(12, 6))

#  Création du graphique avec les données triées
plt.bar(df_sorted['Secteur'], df_sorted['Allocation_Budgetaire (Milliards FCFA)'], 
        color='steelblue')

#  Titres et Labels
plt.title('Budgets par Secteur (Triés par importance)', fontsize=15)
plt.xlabel("secteurs", fontsize=12)
plt.ylabel('Milliards FCFA', fontsize=12)
plt.xticks(rotation=45)

#  Ajouter les valeurs exactes au-dessus de chaque barre
for i, val in enumerate(df_sorted['Allocation_Budgetaire (Milliards FCFA)']):
    plt.text(i, val + 5, str(val), ha='center',fontweight='bold' )

plt.tight_layout()
plt.show()
```

#### 6. Colorez differement le secteur le plus dote.

```python
# Créer une liste de couleurs dynamique
# On met du 'orange' pour le max, 'bleu' pour le reste
couleurs = ['orange' if (val == df_sorted['Allocation_Budgetaire (Milliards FCFA)'].max()) else 'blue' 
            for val in df_sorted['Allocation_Budgetaire (Milliards FCFA)']]

#  Configuration de la figure
plt.figure(figsize=(10, 6))

#  Création du graphique
bars = plt.bar(df_sorted['Secteur'], df_sorted['Allocation_Budgetaire (Milliards FCFA)'], 
               color=couleurs, edgecolor='black')

#  titres
plt.title('Secteur Prioritaire du Budget (Milliards FCFA)', fontsize=14, )
plt.ylabel('Montant (Milliards FCFA)', fontsize=12)
plt.xticks(rotation=45)

#  Ajouter les étiquettes de données sur chaque barre pour plus de clarté
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f"{int(yval)}", 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

# Partie C : Barres horizontales

####  7. Recreez le meme graphique avec des barres horizontales.

```python


#  Trier par ordre CROISSANT pour que le plus grand soit en haut du graphique horizontal
df_sorted_h = df_budget.sort_values(by='Allocation_Budgetaire (Milliards FCFA)', ascending=True)

#  Créer une liste de couleurs (le max en orange, le reste en gris)
couleurs = ['orange' if (val == df_sorted_h['Allocation_Budgetaire (Milliards FCFA)'].max()) else 'blue' 
            for val in df_sorted_h['Allocation_Budgetaire (Milliards FCFA)']]

#  Configuration de la figure
plt.figure(figsize=(10, 6))

#  Utilisation de barh() pour les barres horizontales
bars = plt.barh(df_sorted_h['Secteur'], df_sorted_h['Allocation_Budgetaire (Milliards FCFA)'], 
                color=couleurs,)

#  Habillage
plt.title('Classement des Budgets par Secteur (Milliards FCFA)', fontsize=14, fontweight='bold')
plt.xlabel('Montant (Milliards FCFA)', fontsize=12)
plt.ylabel('Secteurs', fontsize=12)

# Ajouter les étiquettes de données à droite de chaque barre
for bar in bars:
    width = bar.get_width()
    plt.text(width + 5, bar.get_y() + bar.get_height()/2, f'{int(width)}', 
             ha='left', va='center', fontweight='bold')

# Ajuster les marges pour ne pas couper les étiquettes
plt.tight_layout()
plt.xlim(0, 500) # On laisse un peu d'espace à droite pour le texte
plt.show()
```

# Partie E : Barres empilees

##### Decomposez le budget Education en sous-categories

```python

#  Création des données pour les sous-catégories de l'Éducation
donnees_education = {
    'Sous-categorie': ['Primaire', 'Secondaire', 'Superieur', 'Administration'],
    'Budget (Milliards FCFA)': [200, 150, 80, 20]
}

#  Création du DataFrame
df_education = pd.DataFrame(donnees_education)

# Vérification du total et calcul des parts en pourcentage
budget_total_edu = df_education['Budget (Milliards FCFA)'].sum()
df_education['Part (%)'] = ((df_education['Budget (Milliards FCFA)'] / budget_total_edu) * 100).round(2)

# Affichage du résultat
print(f"Budget total de l'Éducation : {budget_total_edu} Milliards FCFA\n")
print(df_education)
```

##### 14. Creez un diagramme a barres empilees pour le secteur Education.

```python
sous_categories = ['Primaire', 'Secondaire', 'Superieur', 'Administration']
budgets = [200, 150, 80, 20]
couleurs = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Couleurs distinctes

#  Configuration de la figure
plt.figure(figsize=(7, 8))

#  Variable pour suivre la hauteur de base (là où la prochaine barre commence)
base = 0

#  Boucle pour construire la barre empilée segment par segment
for i in range(len(sous_categories)):
    budget = budgets[i]
    categorie = sous_categories[i]
    
    # Dessiner le segment
    plt.bar('Éducation', budget, bottom=base, color=couleurs[i], 
            label=categorie, width=0.5, edgecolor='black')
    
    # Ajouter le texte (Montant) au milieu de chaque segment pour plus de clarté
    plt.text(0, base + (budget / 2), f"{categorie}\n{budget} Mds", 
             ha='center', va='center', color='white', fontweight='bold')
    
    # Mettre à jour la base pour le segment suivant
    base += budget

#  Habillage du graphique
plt.title("Décomposition du Budget de l'Éducation (Milliards FCFA)", fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Montant Cumulé (Milliards FCFA)', fontsize=12)

# Placement de la légende en dehors du graphique pour ne pas cacher la barre
plt.legend(title='Sous-catégories', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajuster les marges pour inclure la légende
plt.tight_layout()
plt.show()
```

##### 15. Creez un graphique a 100% empile montrant la repartition proportionnelle.

```python


#  Données
sous_categories = ['Primaire', 'Secondaire', 'Superieur', 'Administration']
budgets = [200, 150, 80, 20]
total_edu = sum(budgets)

#  Conversion en pourcentages (Normalisation à 100%)
parts = [(b / total_edu) * 100 for b in budgets]
couleurs = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Bleu, Orange, Vert, Rouge

#  Création de la figure
plt.figure(figsize=(6, 8))

#  Construction de la barre 100% empilée
base = 0
for i in range(len(sous_categories)):
    part = parts[i]
    plt.bar('Secteur Éducation', part, bottom=base, color=couleurs[i], 
            label=f"{sous_categories[i]}", width=0.6, edgecolor='white', linewidth=1)
    
    # Ajout du texte au centre de chaque segment
    plt.text(0, base + (part / 2), f"{part:.1f}%", 
             ha='center', va='center', color='white', fontweight='bold', fontsize=11)
    
    base += part

#  Finitions esthétiques
plt.title("Structure Proportionnelle du Budget Éducation (100%)", fontsize=13, fontweight='bold', pad=20)
plt.ylabel('Part du budget (%)', fontsize=12)
plt.ylim(0, 100)  # L'axe s'arrête exactement à 100%
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Placement de la légende
plt.legend(title='Sous-catégories', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
