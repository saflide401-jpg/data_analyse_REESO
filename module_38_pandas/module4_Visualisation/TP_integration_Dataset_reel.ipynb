# Exercice 1.3 - Transformation des dates

```python
# 10. Convertissez la colonne date en type datetime
# Convertir en datetime
print("Avec datetime :")
# Convertir en datetime
df['date'] = pd.to_datetime(df['date'])
print("\n Avec datetime :".upper())
print(df)
print("\nType :", df['date'].dtype)
```

```python
# 11. Extrayez l'année, le mois et le jour dans des colonnes séparées
# Extraire les composants
df['Annee'] = df['date'].dt.year
df['Mois'] = df['date'].dt.month
df['Jour'] = df['date'].dt.day

print("\n Extraction de composants :".upper())
print(df)
```

```python
# 12. Identifiez l'année de la première observation et de la dernière
date_debut = df['date'].min()
date_fin = df['date'].max()

print(f"Première observation : {date_debut.year}")
print(f"Dernière observation : {date_fin.year}")
```

```python
# 13. Calculez le nombre d'années couvertes par le dataset
nb_annees = date_fin.year - date_debut.year
print(f"Le dataset couvre {nb_annees} années.")
```
