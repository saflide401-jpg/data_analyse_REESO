##### 2. Creez un DataFrame Mouvements_stock avec l'historique sur 6 mois :

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. Configuration des données de base
produits = {
    'P001': 'Paracetamol', 'P002': 'Amoxicilline', 
    'P003': 'Quinine', 'P004': 'SRO', 'P005': 'Artemether'
}
depots = ['Central', 'Ouaga', 'Bobo', 'Ouahigouya']
motifs_entree = ['Achat']
motifs_sortie = ['Vente', 'Peremption', 'Casse']

# 2. Génération des dates sur 6 mois
date_fin = datetime.now()
dates = [date_fin - timedelta(days=x) for x in range(180)]

# 3. Création des mouvements (simulation de 1000 lignes)
n_lignes = 1000
data = {
    'Date': np.random.choice(dates, n_lignes),
    'Code_produit': np.random.choice(list(produits.keys()), n_lignes),
    'Type': np.random.choice(['Entree', 'Sortie'], n_lignes, p=[0.3, 0.7]),
    'Depot': np.random.choice(depots, n_lignes),
    'Quantite': np.random.randint(1, 500, n_lignes)
}

Mouvements_stock = pd.DataFrame(data)

# 4. Attribution cohérente du Motif selon le Type
def attribuer_motif(row):
    if row['Type'] == 'Entree':
        return 'Achat'
    else:
        return np.random.choice(motifs_sortie, p=[0.85, 0.10, 0.05])

Mouvements_stock['Motif'] = Mouvements_stock.apply(attribuer_motif, axis=1)

# Tri chronologique
Mouvements_stock = Mouvements_stock.sort_values('Date').reset_index(drop=True)

print(Mouvements_stock.head())
```

# Partie B : Calcul des stocks
4. Calculez le stock actuel de chaque produit par depot.

```python


# Produits
produits = ["PR001", "PR002", "PR003", "PR004"]

# Dates sur 6 mois (tous les 15 jours)
dates = pd.date_range("2025-09-01", "2026-02-28", freq="15D")

# Création du tableau directement
df_mouvements = pd.DataFrame({
    "Date": np.repeat(dates, len(produits)),          # chaque date répétée pour chaque produit
    "Code_produit": np.tile(produits, len(dates)),    # chaque produit répété pour toutes les dates
    "Type": np.random.choice(["Entree", "Sortie"], size=len(dates)*len(produits)),
    "Quantite": np.random.randint(10, 200, size=len(dates)*len(produits)),
    "Motif": np.random.choice(["Achat", "Vente", "Peremption", "Casse"], size=len(dates)*len(produits)),
    "Depot": np.random.choice(["Central", "Ouaga", "Bobo", "Ouahigouya"], size=len(dates)*len(produits))
})

print(df_mouvements.head(10))
```

# Partie B : Calcul des stocks
4. Calculez le stock actuel de chaque produit par depot.

Le stock actuel se calcule comme :

Stock actuel  =   Somme des Entrees − Somme des Sorties

```python

# Séparer entrées et sorties
df_mouvements["Quantite_signée"] = df_mouvements.apply(
    lambda x: x["Quantite"] if x["Type"] == "Entree" else -x["Quantite"], axis=1
)

# Stock actuel par produit et dépôt
stock_actuel = df_mouvements.groupby(["Code_produit", "Depot"])["Quantite_signée"].sum().reset_index()
print(stock_actuel)
```

##### 4-Consommation Moyenne Mensuelle (CMM)

La CMM correspond à la moyenne des sorties mensuelles (ventes, péremptions, casses).


```python
# CMM= Total des sorties sur la periode/Nombre de mois

# Filtrer uniquement les sorties
sorties = df_mouvements[df_mouvements["Type"] == "Sortie"]

# Regrouper par produit et mois
sorties["Mois"] = sorties["Date"].dt.to_period("M")
cmm = sorties.groupby(["Code_produit", "Mois"])["Quantite"].sum().groupby("Code_produit").mean().reset_index()
cmm.rename(columns={"Quantite": "CMM"}, inplace=True)
print(cmm)
```

# 6 Stock de sécurité


Le stock de sécurité est défini comme 2 mois de CMM :

```python
cmm["Stock_securite"] = 2 * cmm["CMM"]
```

# Niveau de commande
Le niveau de commande combine le stock de sécurité et le délai de réapprovisionnement exprimé en jours :

Niveau de commande = Stock securite + (CMM journaliere × Delai reapprovisionnement)

```python
# Joindre avec le DataFrame Produits pour récupérer le délai
df_commande = cmm.merge(produits[["Code_produit", "Delai_reapprovisionnement_jours"]], on="Code_produit")

# Calcul du niveau de commande
df_commande["CMM_journaliere"] = df_commande["CMM"] / 30
df_commande["Niveau_commande"] = df_commande["Stock_securite"] + (
    df_commande["CMM_journaliere"] * df_commande["Delai_reapprovisionnement_jours"]
)

print(df_commande)
```

# partie C alertess

```python
# Fusionner stock actuel avec les indicateurs
df_alertes = stock_actuel.merge(df_commande[["Code_produit","CMM","Stock_securite","Niveau_commande"]], on="Code_produit")

# Classification
def classification(row):
    if row["Quantite_signée"] == 0:
        return "Rupture de stock"
    elif row["Quantite_signée"] < row["Stock_securite"]:
        return "Alerte critique"
    elif row["Quantite_signée"] < row["Niveau_commande"]:
        return "Alerte"
    elif row["Quantite_signée"] > 12 * row["CMM"]:
        return "Surstock"
    else:
        return "Stock correct"

df_alertes["Statut"] = df_alertes.apply(classification, axis=1)
print(df_alertes)
```

# 9 Rapport d’alerte classé par priorité

```python
priorite = {"Rupture de stock":1, "Alerte critique":2, "Alerte":3, "Stock correct":4, "Surstock":5}
df_alertes["Priorite"] = df_alertes["Statut"].map(priorite)

rapport = df_alertes.sort_values("Priorite")
print(rapport)
```

# 10 Valeur financière des stocks par dépôt

```python
# Joindre avec prix unitaire
df_valeur = stock_actuel.merge(produits[["Code_produit","Prix_unitaire_FCFA"]], on="Code_produit")

# Calcul valeur
df_valeur["Valeur_stock_FCFA"] = df_valeur["Quantite_signée"] * df_valeur["Prix_unitaire_FCFA"]

# Valeur par dépôt
valeur_par_depot = df_valeur.groupby("Depot")["Valeur_stock_FCFA"].sum().reset_index()
print(valeur_par_depot)
```

 # Partie D : Analyse des mouvements

```python
# 11. Analysez les sorties par mois et par produit pour detecter les tendances.

# Filtrer uniquement les sorties
sorties = df_mouvements[df_mouvements["Type"] == "Sortie"].copy()

# Ajouter une colonne Mois
sorties["Mois"] = sorties["Date"].dt.to_period("M")

# Agréger par produit et mois
sorties_mensuelles = sorties.groupby(["Code_produit", "Mois"])["Quantite"].sum().reset_index()

print(sorties_mensuelles.head(10))
```

```python
# 12-Produits avec les plus fortes variations de consommation
variation = sorties.groupby(["Code_produit", "Mois"])["Quantite"].sum().groupby("Code_produit").agg(["mean","std","max","min"])
variation["Amplitude"] = variation["max"] - variation["min"]

print(variation)
```

```python
# 13 Coût des pertes (péremptions + casse)
# Filtrer pertes
pertes = sorties[sorties["Motif"].isin(["Peremption","Casse"])]

# Joindre avec prix unitaire
pertes = pertes.merge(produits[["Code_produit","Prix_unitaire_FCFA"]], on="Code_produit")

# Calcul coût des pertes
pertes["Cout_perte"] = pertes["Quantite"] * pertes["Prix_unitaire_FCFA"]

# Total par produit
cout_pertes = pertes.groupby("Code_produit")["Cout_perte"].sum().reset_index()

# Total global
cout_total = pertes["Cout_perte"].sum()

print(cout_pertes)
print("Coût total des pertes sur la période :", cout_total, "FCFA")
```

# 14 Estimation des besoins pour les 3 prochains mois
On utilise les sorties mensuelles par produit (calculées en Partie D) et on applique une projection simple :



```python
# Besoin prévisionnel simple = 3 mois de CMM
previsions = cmm.copy()
previsions["Besoin_3mois"] = previsions["CMM"] * 3
print(previsions)
```

# 14 Génération des commandes nécessaires

Pour éviter les ruptures, on compare le stock actuel avec le niveau de commande et les besoins futurs :

Si stock actuel < niveau de commande → déclencher une commande.

Quantité à commander = Besoin 3 mois + Stock sécurité – Stock actuel.

On tient compte du délai de réapprovisionnement pour anticiper.

```python
# Fusionner stock actuel avec prévisions
df_commande_finale = stock_actuel.merge(previsions[["Code_produit","Besoin_3mois","Stock_securite"]], on="Code_produit")

# Joindre délai
df_commande_finale = df_commande_finale.merge(produits[["Code_produit","Delai_reapprovisionnement_jours"]], on="Code_produit")

# Calcul commande
df_commande_finale["Commande"] = (
    df_commande_finale["Besoin_3mois"] + df_commande_finale["Stock_securite"] - df_commande_finale["Quantite_signée"]
).clip(lower=0)

print(df_commande_finale)
```
