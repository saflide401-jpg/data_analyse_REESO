# TP-14 : Gestion des stocks – CAMEG (Solution Python)

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

# Partie A : Données de stock
# 1. DataFrame Produits

```python
produits = pd.DataFrame({
    "Code_produit": ["P001","P002","P003","P004","P005"],
    "Designation": ["Paracetamol","Amoxicilline","Quinine","SRO","Ceftriaxone"],
    "Forme": ["Comprime","Capsule","Injectable","Sachet","Injectable"],
    "DCI": ["Paracetamol","Amoxicilline","Quinine","Sels de rehydratation","Ceftriaxone"],
    "Dosage": ["500mg","500mg","300mg","20.5g","1g"],
    "Unite": ["Boite","Boite","Ampoule","Sachet","Flacon"],
    "Prix_unitaire_FCFA": [500,1200,800,300,2500],
    "Fournisseur_principal": ["Lab A","Lab B","Lab C","Lab D","Lab B"],
    "Delai_reapprovisionnement_jours": [30,45,40,20,50]
})
print (produits)
```

# 2. DataFrame Mouvements_stock

```python
dates = pd.date_range(end=datetime.today(), periods=180)
print(dates)

```

```python
mouvements = []

depots = ["Central","Ouaga","Bobo","Ouahigouya"]
motifs = ["Achat","Vente","Peremption","Casse"]
vendeur=["Francis","alizata", "Aminata","Delfine", "Diane", "Siaka", "Latif"]


for d in dates:
    for code in produits["Code_produit"]:
        mouvements.append([
            d,
            code,
            np.random.choice(["Entree","Sortie"]),
            np.random.randint(10,200),
            np.random.choice(motifs),
            np.random.choice(depots),
            np.random.choice(vendeur)
        ])

mouvements_stock = pd.DataFrame(
    mouvements,
    columns=["Date","Code_produit","Type","Quantite","Motif","Depot", "vendeur"]
)

print("les mouvements du stock")
print(mouvements_stock.head())
```

```python

print("les informations sur le dataaframe mouvements_stocks\n".upper())
mouvements_stock.info()

print("\n la forme est:")

mouvements_stock.shape()
```

# Partie B : Calcul des stocks
# 4. Stock actuel par produit et dépôt

```python
mouvements_stock["Quantite_calc"] = np.where(
    mouvements_stock["Type"]=="Entree",
    mouvements_stock["Quantite"],
    -mouvements_stock["Quantite"]
)

stock_actuel = mouvements_stock.groupby(
    ["Code_produit","Depot"]
)["Quantite_calc"].sum().reset_index()

stock_actuel.rename(columns={"Quantite_calc":"Stock_actuel"}, inplace=True)
print(stock_actuel)
```

```python
# On multiplie la quantité par 1 si Entrée, sinon par -1
mouvements_stock["Stock_actuel"] = mouvements_stock["Quantite"] * mouvements_stock["Type"].map({"Entree": 1, "Sortie": -1})

stock_actuel = mouvements_stock.groupby(["Code_produit", "Depot"])["Stock_actuel"].sum().reset_index()
print("\n le stock actuel")
print(stock_actuel)
```

```python
# usage du pivot 
pivot = mouvements_stock.pivot_table(
    index=["Code_produit", "Depot"], 
    columns="Type", 
    values="Quantite", 
    aggfunc="sum", 
    fill_value=0
)

# Stock = Entrées - Sorties
pivot["Stock_actuel"] = pivot["Entree"] - pivot.get("Sortie", 0)
stock_actuel = pivot["Stock_actuel"].reset_index()
print(stock_actuel)
```

# 5. Consommation Moyenne Mensuelle (CMM)

```python
sorties = mouvements_stock[mouvements_stock["Type"]=="Sortie"]

cmm = sorties.groupby("Code_produit")["Quantite"].sum() / 6
cmm = cmm.reset_index()
cmm.rename(columns={"Quantite":"CMM"}, inplace=True)

print("\n les mouvements de sorties".upper())
print(sorties)

print ("\n la commande moyennemensuelle:")
print(cmm)
```

# 6. Stock de sécurité

```python
cmm["Stock_securite"] = 2 * cmm["CMM"]
print(cmm)
```

# 7. Niveau de commande

```python
produits_cmm = produits.merge(cmm, on="Code_produit")

produits_cmm["consommation_moyenne_journaliere"] = produits_cmm["CMM"] / 30

produits_cmm["Niveau_commande"] = (
    produits_cmm["Stock_securite"] +
    produits_cmm["Delai_reapprovisionnement_jours"] *
    produits_cmm["consommation_moyenne_journaliere"]
)
print(produits_cmm)
```

# Partie C : Alertes de stock
# 8. Classification des stocks

```python
stock = stock_actuel.merge(produits_cmm, on="Code_produit")

def statut_stock(row):
    
    if row["Stock_actuel"] == 0:
        return "Rupture"
    
    elif row["Stock_actuel"] < row["Stock_securite"]:
        return "Alerte critique"

    elif row["Stock_actuel"] < row["Niveau_commande"]:
        return "Alerte"
    
    elif row["Stock_actuel"] > 12 * row["CMM"]:
        return "Surstock"
    
    else:
        return "Stock correct"
    print("stçck classifier")

stock["Statut"] = stock.apply(statut_stock, axis=1)
print(stock_actuel)
```

# 9. Rapport d’alerte

```python
priorite = {
    "Rupture":1,
    "Alerte critique":2,
    "Alerte":3,
    "Stock correct":4,
    "Surstock":5
}

stock["Priorite"] = stock["Statut"].map(priorite)

rapport_alerte = stock.sort_values("Priorite")
print("\n rapport d'rapport_alerte")
print(rapport_alerte)
```

# 10. Valeur financière des stocks

```python
stock["Valeur_stock"] = stock["Stock_actuel"] * stock["Prix_unitaire_FCFA"]
print("valeur des stocks")
valeur_depot = stock.groupby("Depot")["Valeur_stock"].sum()

print("valeur des stocks")
print(valeur_depot)
```

# Partie D : Analyse des mouvements
# 11. Sorties par mois et produit

```python
sorties["Mois"] = sorties["Date"].dt.to_period("M")

analyse_mensuelle = sorties.groupby(
    ["Mois","Code_produit"]
)["Quantite"].sum().reset_index()

print("sortie par mois et produits")
print(analyse_mensuelle)
```

# 12. Produits avec fortes variations

```python
variation = analyse_mensuelle.groupby("Code_produit")["Quantite"].std()

variation = variation.sort_values(ascending=False)
print("produits avec forte variation")
print(variation)
```

# 13. Coût des pertes

```python
pertes = mouvements_stock[mouvements_stock["Motif"].isin(["Peremption","Casse"])]

pertes = pertes.merge(produits[["Code_produit","Prix_unitaire_FCFA"]], on="Code_produit")

pertes["Cout"] = pertes["Quantite"] * pertes["Prix_unitaire_FCFA"]

cout_total_pertes = pertes["Cout"].sum()
print("\n les pertes")
print(pertes)
print("\n le cout total des pertes")
print(cout_total_pertes)
```

# Partie E : Prévisions
# 14. Besoins pour les 3 prochains mois

```python
previsions = cmm.copy()

previsions["Besoin_3_mois"] = 3 * previsions["CMM"]
print("\n dans 3mois")
print(previsions)
```

# 15. Génération des commandes

```python
commandes = stock.merge(previsions, on="Code_produit")

commandes["Quantite_commande"] = (
commandes["Besoin_3_mois"] + commandes["Stock_securite"] - commandes["Stock_actuel"])

commandes["Quantite_commande"] = commandes["Quantite_commande"].apply(
    lambda x: max(0, x))

commandes_finales = commandes[commandes["Quantite_commande"] > 0][
    ["Code_produit","Depot","Quantite_commande"]
]
```
