Objectif : Créer un tableau de bord interactif avec Plotly.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
```

```python
print("=== DASHBOARD INTERACTIF - ANALYSE COMMERCIALE ===\n")
```

```python
# 1. Générer données
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=365, freq='D')
n_ventes = 1000
print(n_ventes)
```

```python
df = pd.DataFrame({
'Date': np.random.choice(dates, n_ventes),
'Produit': np.random.choice(['Mil', 'Riz', 'Maïs', 'Sorgho'], n_ventes),
'Région': np.random.choice(['Centre', 'Hauts-Bassins', 'Nord'], n_ventes),
'Quantité': np.random.randint(10, 100, n_ventes),
'Prix_Unitaire': np.random.uniform(300, 1000, n_ventes)})
print(df)
```

```python
df['Montant'] = df['Quantité'] * df['Prix_Unitaire']
df['Mois'] = df['Date'].dt.month
df['Trimestre'] = df['Date'].dt.quarter
```

```python
print("1. DONNÉES GÉNÉRÉES")
print(df.head())
print(f"\nNombre de transactions : {len(df)}")
print(f"CA total : {df['Montant'].sum():,.0f} FCFA")
```

creer un dashbord


```python
# 2. Créer dashboard avec subplots
fig = make_subplots(
rows=2, cols=2,
subplot_titles=('CA mensuel', 'CA par produit',

'Distribution montants', 'CA par région'),

specs=[[{'type': 'scatter'}, {'type': 'bar'}],
[{'type': 'histogram'}, {'type': 'pie'}]])
```

```python
# Graphique 1 : Évolution mensuelle
ca_mensuel = df.groupby('Mois')['Montant'].sum().reset_index()
fig.add_trace(
go.Scatter(x=ca_mensuel['Mois'], y=ca_mensuel['Montant'],

mode='lines+markers', name='CA mensuel',
line=dict(color='blue', width=3),
marker=dict(size=8)),

row=1, col=1
)
```
